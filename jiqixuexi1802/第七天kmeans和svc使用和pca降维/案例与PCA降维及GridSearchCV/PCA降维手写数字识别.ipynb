### 手写数字识别

```python
import sklearn.datasets as datasets
```

```python
# 主成分分析
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()

X = iris['data']
y = iris['target']

X
```

```python
X.shape
```

```python
# whiten,白化操作-----归一化操作（Z-score）
pca = PCA(n_components=2,whiten = True)

pca.fit(X)

X_pca = pca.transform(X)
X_pca.shape
```

```python
X
```

```python
# ? ? ? 矩阵运算
# 求解了特征值（矩阵特征），特征向量
X_pca
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf')

svc.fit(X,y)

svc.score(X,y)
```

```python
svc = SVC(kernel='rbf')

svc.fit(X_pca,y)

svc.score(X_pca,y)
```

```python
# 第一步求解X平均值
# 4个属性的平均值
X_mean = X.mean(axis = 0)
X_mean
```

```python
# 第二步相减（特征中心化）
C = X - X_mean
C
```

```python
import numpy as np
```

```python
# 第三步求解协方差
# Estimate a covariance matrix, given data and weights
B = np.cov(C,rowvar=False)
B
```

```python
# 第四步，求解协方差的特征值和特征向量
np.linalg.eigvals(B)
```

```python
eig,eigvector = np.linalg.eig(B)
```

```python
X_pca
```

```python
# 第五步，将数据进行降维
# X.shape = (150,4) ---->前两个最大特征值对应特征向量对数据进行降维
ret = X.dot(eigvector[:,:2])
```

```python
X_pca
```

```python
# 第六步，z-score归一化
(ret - ret.mean(axis = 0))/ret.std(axis = 0)
```

```python
eig
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
eig
```

```python
pca = PCA(n_components=0.9999,whiten=True)

pca.fit(X)

X_pca = pca.transform(X)
display(X_pca.shape,pca.explained_variance_,pca.explained_variance_ratio_)
```

```python
X_train = X.iloc[:5000,:]

y_train = y[:5000]
display(X_train.shape,y_train.shape)
```

```python
X_test = X.iloc[-1000:,:]
y_test = y[-1000:]
```

```python
%%time
# 没有经过数据降维
svc = SVC(kernel='rbf')

# 小样本数据，数据量不能太大
svc.fit(X_train,y_train)

print(svc.score(X_test,y_test))
```

```python
# 使用PCA进行降维X_train.shape = (5000,784)

pca = PCA(n_components=0.95,whiten=True)

pca.fit(X_train)

X_train_pca = pca.transform(X_train)

X_test_pca = pca.transform(X_test)

# 原来784属性降维到149个
X_train_pca.shape
```

```python
%%time
svc = SVC(kernel='rbf')

svc.fit(X_train_pca,y_train)

# 预测
print('经过PCA降维，准确率：',svc.score(X_test_pca,y_test))
```
