```python
import sklearn.datasets as datasets
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = datasets.load_iris()
iris
```

```python
#data 数据， 对应 target
x = iris['data']
y = iris['target']
display(x,y)
```

```python
import numpy as np 
index = np.arange(150)
#打乱index
np.random.shuffle(index)
index
```

```python
#打乱数据去 rins里面去取
X = x[index]
Y = y[index]
# display(X,Y)
```

```python
# #拿一部分测试数据
# x_test = X[105:]
# y_test = y[105:]
# #拿一部分训练数据  7 ：3 比例
# x_train = x[:105]
# y_train = y[:105]

#拿一部分训练数据  7 ：3 比例
x_train = X[:105]
y_train = Y[:105]
#拿一部分测试数据
x_test = X[105:]
y_test = Y[105:]
# x_train
```

```python
# knn = KNeighborsClassifier(n_neighbors=5)
# knn.fit(x_train,y_train)
# y_ = knn.predict(x_test)
# display(y_,y_test)
# conn = y_ ==y_test
# conn
# conn.mean()
```

```python
#权重 distance  ，不够精准 但是更加真实了。knn 是监督学习， knn分类，归一化。  p =1  曼哈顿距离 p=2 欧几里得距离 
knn = KNeighborsClassifier(n_neighbors=5,weights='distance',p=1)
#训练
knn.fit(x_train,y_train)
#测试
knn.score(x_test,y_test)
```
