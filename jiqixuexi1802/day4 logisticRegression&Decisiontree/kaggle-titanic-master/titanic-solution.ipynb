### Bar Chart for Categorical Features
- Pclass
- Sex
- SibSp ( # of siblings and spouse)
- Parch ( # of parents and children)
- Embarked
- Cabin

```python
def bar_chart(feature):
    #train['Survived']==1 来判断存活人，把这个作为条件传入train，然后传入一个特征，统计这个特征存活总数
    survived = train[train['Survived']==1][feature].value_counts()
    #train['Survived']==1 来判断死亡人，把这个作为条件传入train，然后传入一个特征，统计这个特征死亡总数
    dead = train[train['Survived']==0][feature].value_counts()
    #把存活数据和死亡数据存放到dataframe中
    df = pd.DataFrame([survived,dead])
    #设置索引
    df.index = ['Survived','Dead']
    #df自带画图功能，kind是柱状图，figsize画布大小  生成堆叠条形图的haunted，可以设置参数 stacked=True 你可以设置False看看效果
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

The Chart confirms **a person aboarded with more than 2 siblings or spouse** more likely survived

### 4.2 Name

```python
from pandas import Series
#把测试数据和测试数据放在一个列表中，这个list长度为2 第一个train，第二个是test 每个元素是dataframe不信你看type(train_test_data[0])
train_test_data = [train, test] # combining train and test dataset
for dataset in train_test_data:
    #给每个df数据集增加一个title列 获取规则是名字那列值根据正则截取空格和.之间全字母数据 如Braund, Mr. Owen Harris 得到Mr
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

#str 只针对Series expand=True返回dataframe。expand=False返回Series/Index/DataFrame，如果只匹配到一列返回series，匹配不到返回index，匹配多列dataframe。默认Fals
# a = Series(["Moran, Mr.","Moran, ZH."])
# print(a.str.extract(' ([A-Za-z]+)\.', expand=False))
```

```python
#统计训练数据总数
train['Title'].value_counts()
```

```python
#统计测试数据总数
test['Title'].value_counts()
```

#### Title map
Mr : 0  
Miss : 1  
Mrs: 2  
Others: 3


```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
#利用map函数把每个title从字符串转化成数字，方便计算和节省内存我猜
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].map(title_mapping)
```

#### 4.4.2 Binning
Binning/Converting Numerical Age to Categorical Variable  

feature vector map:  
child: 0  
young: 1  
adult: 2  
mid-age: 3  
senior: 4

```python
for dataset in train_test_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0,
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1,
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2,
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3,
    dataset.loc[ dataset['Age'] > 62, 'Age'] = 4
```

## 5. Modelling

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

```python
train.info()
```

### 6.2 Cross Validation (K-fold)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

### 6.2.1 kNN

```python
clf = KNeighborsClassifier(n_neighbors = 13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# kNN Score
round(np.mean(score)*100, 2)
```

### 6.2.2 Decision Tree

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# decision tree Score
round(np.mean(score)*100, 2)
```

### 6.2.3 Ramdom Forest

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Random Forest Score
round(np.mean(score)*100, 2)
```

### 6.2.4 Naive Bayes

```python
clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Naive Bayes Score
round(np.mean(score)*100, 2)
```

### 6.2.5 SVM

```python
clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

## 7. Testing

```python
clf = SVC()
clf.fit(train_data, target)

test_data = test.drop("PassengerId", axis=1).copy()
prediction = clf.predict(test_data)
```

```python
submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": prediction
    })

submission.to_csv('submission.csv', index=False)
```
